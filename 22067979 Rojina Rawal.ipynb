```python
#Data Preparation

#Q.no.1: Python program to load data into pandas DataFrame.

#Importing the pandas and numpy library
import pandas as pd
import numpy as np

#Loading the dataset using the pd.read_csv() function.
data = pd.read_csv('DataScienceSalaries_8b290669-a5e9-45bf-be72-d27add2eacae_93472_.csv')
data
```

```python
#Q.no.2: Python program to remove unnecessary columns i.e., salary and salary currency. 

# Dropping the 'salary' and 'salary_currency' columns from data
data = data.drop(columns=['salary', 'salary_currency'])
data
print("\nDataFrame after removing 'salary' and 'salary_currency' columns:")
data
```

```python
# Check for missing values in the DataFrame
data.isna()
```

```python
#Q.no.3: Python program to remove the NaN missing values from updated dataframe.
    
# Since there are no NaN missing values in the DataFrame, nothing will be dropped.
data = data.dropna()
data
```

```python
#Q.no.4: Python program to check duplicates value in the dataframe. 

#using the duplicated() function to check for duplicate rows 
data[data.duplicated()]
```

```python
# dictionary to store unique values for each column
unique_column_values = {}

# Using for each loop
for column in data.columns:
    #unique values for the current column
    unique_values = data[column].unique()
    # Storing unique values with column name as the key
    unique_column_values[column] = unique_values

# Printing the dictionary
for column, values in unique_column_values.items():
    print(f"Unique values in column '{column}': {values}")
```

```python
# Dictionary to map Machine Learning Engineer to ML Engineer
rename_similar_job_title = {'Machine Learning Engineer': 'ML Engineer'}

# using inplace= True to replace the original data frame
data['job_title'].replace(rename_similar_job_title, inplace=True)

# dictionary to store unique values for each column
unique_column_values = {}

# Using for each loop
for column in data.columns:
    #unique values for the current column
    unique_values = data[column].unique()
    # Storing unique values with column name as the key
    unique_column_values[column] = unique_values

# Printing the dictionary
for column, values in unique_column_values.items():
    print(f"Unique values in column '{column}': {values}")
```

```python
# Qno.6 Rename the experience level SE â€“ Senior Level/Expert, MI â€“ Medium Level/Intermediate, EN â€“ Entry Level,EX â€“ Executive Level

# storing the values in dictionary
rename_values={'SE': 'Senior Level/Expert',
               'MI': 'Medium Level/Intermediate',
               'EN': 'Entry Level',
               'EX': 'Executive Level'}

#using replace() function to map old values to new values and inplace=True modifies the original DataFrame
data['experience_level'].replace(rename_values, inplace= True)
data
```

```python
# Data Analysis
#Qno.1 Python program to show summary statistics of sum, mean, standard deviation, skewness, and kurtosis of any chosen variable.

#Calculating the sum of 'salary_in_usd' column
sum= data['salary_in_usd'].sum()
print("Sum of salary in USD:",sum)
```

```python
#Calculating mean using .mean() function
mean_salary = data['salary_in_usd'].mean()
print("Mean salary in USD:", mean_salary)
```

```python
#Calculating standard deviation using .std() function
sd= data['salary_in_usd'].std()
print("Standard deviation of salary in USD:", sd)
```

```python
# importing skew and kurtosis function from scipy.stats
from scipy.stats import skew, kurtosis

# calculating skewness of 'salary_in_usd' column
skewness = skew(data['salary_in_usd'])

# Print the calculated skewness
print("Skewness of 'salary_in_usd' column:", skewness)
```

```python
# kurtosis of 'salary_in_usd' column
salary_kurtosis = kurtosis(data['salary_in_usd'])

# Print the calculated kurtosis
print("Kurtosis of 'salary_in_usd' column:", salary_kurtosis)
```

```python
# Q.no.2: Python program to calculate and show correlation of all variables.

# Correlation of the numeric columns in the DataFrame
correlation = data.corr(numeric_only=True)
print("Correlation of all variables:")
correlation
```

```python
#Data Exploration

#Q.no.1: Python program to find out top 15 jobs. Make a bar graph of sales as well.   

# importing the matplotlib.pyplot library for data visualization
import matplotlib.pyplot as plt

# calculating the 15 most common job titles and their frequencies
top_jobs = data['job_title'].value_counts().head(15)

# using .plot() function to plot bar graph 
top_jobs.plot(kind='bar', color='pink')

# Setting the labels and title
plt.xlabel('Job Title')
plt.ylabel('Frequency')
plt.title('Top 15 Jobs')
```

```python
#Q.no.2: Which job has the highest salaries? Illustrate with bar graph.

# Calculate the mean salary for each job title
average_salary_foreach_job = data.groupby('job_title')['salary_in_usd'].mean().head(10)

# Find the job with the highest mean salary
highest_salary_job = average_salary_foreach_job.idxmax()

# Create a bar graph to visualize the mean salary for each job title
average_salary_foreach_job.plot(kind='bar')  # Plotting a bar graph
plt.xlabel('Job Title')  # Label for x-axis
plt.ylabel('Mean Salary')  # Label for y-axis
plt.title('Mean Salary by Job Title')  # Title of the plot
plt.axhline(average_salary_foreach_job[highest_salary_job], color='red', linestyle='--', label=f'Highest Salary Job: {highest_salary_job}')  # Horizontal line indicating the highest salary job
plt.legend()  # Show legend
plt.show()  # Display the plot
```

```python
#Q.no.3: python program to find out salaries based on experience level. Illustrate it through bar graph.

# Grouping the data by 'experience_level' and calculating the mean salary
salary_by_experience = data.groupby('experience_level')['salary_in_usd'].mean()

salary_by_experience.plot(kind='bar', color='lightpink', edgecolor='black')  # Plotting a bar graph with specified colors and edge color
plt.xlabel('Experience Level')  # Label for x-axis
plt.ylabel('Mean Salary')  # Label for y-axis
plt.title('Mean Salary by Experience Level')  # Title of the plot   
```

```python
# Q.no.4: Python program to show histogram and box plot of any chosen different variables. Use proper labels in the graph.


chosen_variable = 'salary_in_usd' 
plt.figure(figsize=(10, 5))  # Setting the figure size
plt.hist(data[chosen_variable], bins=20, color='lightpink', edgecolor='black')  # Plotting histogram with specified colors and edge color
plt.title('Histogram of ' + chosen_variable)  # Title of the histogram
plt.xlabel(chosen_variable)  # Label for x-axis
plt.ylabel('Frequency')  # Label for y-axis

#box plot
plt.figure(figsize=(10, 5))  # Set the figure size
plt.boxplot(data[chosen_variable], vert=False)  # Plot box plot horizontally
plt.title('Boxplot of ' + chosen_variable)  # Title of the box plot
plt.grid(True)
plt.xlabel(chosen_variable)  # Label for x-axis
```
